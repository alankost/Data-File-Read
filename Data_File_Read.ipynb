```python
import os
from tkinter import *
from tkinter import filedialog
import pandas as pd
import glob
from openpyxl import Workbook, load_workbook
from openpyxl.chart import (LineChart,Reference)

fold_path = "unassigned"

def Get_Folder_Path():
    global folder_path
    folder_path = filedialog.askdirectory(initialdir=os.path.normpath("C:"), title="Example")
    root.destroy()

# if __name__ == '__main__':
    
root=Tk()
root.geometry("200x100")
root.attributes("-topmost", True)
b = Button(root, text='Select Folder', command = Get_Folder_Path).pack(fill=X)
root.mainloop()

li = []

for file in os.listdir(folder_path):
    if os.path.splitext(file)[1] == '.csv':
        f = os.path.join(folder_path, file)
        df = pd.read_csv(f, index_col=None, header=0)
        li.append(df)

frame = pd.concat(li, axis=1, ignore_index=True)

from openpyxl.utils.dataframe import dataframe_to_rows
wb = Workbook()
ws = wb.active
ws.title = "Data"

for r in dataframe_to_rows(frame, index=False, header=False):
    ws.append(r)

from openpyxl.utils import get_column_letter
mc = ws.max_column
mr = ws.max_row
get_column_letter(mc)+str(mr)


for x in ws['B5':get_column_letter(mc)+str(mr)]:
    for y in x:
        y.value=float(y.value)
        
li = []

for file in os.listdir(folder_path):
    if os.path.splitext(file)[1] == '.csv':
        f = os.path.join(folder_path, file)
        df = pd.read_csv(f, index_col=None, header=0)
        li.append(df)

frame = pd.concat(li, axis=1, ignore_index=True)

from openpyxl.utils.dataframe import dataframe_to_rows
wb = Workbook()
ws = wb.active
ws.title = "Data"

for r in dataframe_to_rows(frame, index=False, header=False):
    ws.append(r)

from openpyxl.utils import get_column_letter
mc = ws.max_column
mr = ws.max_row
get_column_letter(mc)+str(mr)

for x in ws['B5':get_column_letter(mc)+str(mr)]:
    for y in x:
        y.value=float(y.value)

csv_files=[]
for file in os.listdir(folder_path):
    if os.path.splitext(file)[1] == '.csv':
        csv_files.append(file)
k=0
for file in csv_files:
        xvalues=Reference(ws, min_row = 5, min_col = 1+k*8, 
                            max_row = mr)
        yvalues = Reference(ws, min_row = 5, min_col = 2+k*8, 
                                max_row = 4000)
        chart = LineChart()
        chart.x_axis.title = 'Position (micron)'
        chart.y_axis.title = 'Relative Index'
        chart.set_categories(xvalues)
        chart.add_data(yvalues)
        chart.set_categories(xvalues)
        wb.create_chartsheet(title=file[0:8]).add_chart(chart)
        k+=1
        filename=csv_files[0][0:7]+'-'+csv_files[-1][0:7]+'.xlsx'
wb.save(os.path.join(folder_path, filename))
```
